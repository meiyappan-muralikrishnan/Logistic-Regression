### SCALING

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

wine_quality.iloc[:, 0:13] = scale.fit_transform(wine_quality.iloc[:, 0:13])
```

### CATEGORICAL VALUES

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

wine_quality["Wine Type"] = le.fit_transform(wine_quality["Wine Type"])
```

```python
wine_quality["Wine Type"].value_counts()
```

```python
correlation = wine_quality.corr()

correlation
```

```python
correlation[(correlation["Wine Type"].abs())>0.1].columns
```

```python
x = wine_quality[['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
       'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
       'Proanthocyanins', 'Color intensity', 'Hue',
       'OD280/OD315 of diluted wines', 'Proline']]

y = wine_quality["Wine Type"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=100)
```

```python
y_train.value_counts()
```

### BALANCING THE DATASET

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(sampling_strategy="minority", random_state=100)

x_train_os, y_train_os = sm.fit_sample(x_train, y_train)
```

```python
y_train_os.value_counts()
```

```python
# Multiclass logistic regression, [multi_class="ovr"],not mandatory inorder to safer side, provide like this

# OVR - One vs Rest or One vs All

from sklearn.linear_model import LogisticRegression

log = LogisticRegression(multi_class="ovr")

log.fit(x_train_os, y_train_os)
```

```python
log.predict(x_test)
```

```python
log.predict_log_proba(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test, log.predict(x_test))
```

```python
print(metrics.classification_report(y_test, log.predict(x_test)))
```

```python
# need to verify whether log loss function can be verified for Multi Class Classification

metrics.log_loss(y_test, log.predict_proba(x_test))
```
